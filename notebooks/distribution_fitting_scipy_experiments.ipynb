```python
import pandas as pd
import fitter
import pickle as pkl
```

# Level 1 EV Charging

The start times and durations are fitted to the data and then the truncated distributions are fit to see if there are any appreciable differnces in the fits.

```python
#raw start times
l1_st = pd.read_csv('../data/L1_start_times.csv')
f_lev1_st = fitter.Fitter(l1_st)
f_lev1_st.fit()
# may take some time since by default, all distributions are tried
# but you call manually provide a smaller set of distributions
f_lev1_st.summary()
```

```python
#raw durations
l1_dur = pd.read_csv('../data/l1_durations_xformed.csv')
f1dur = fitter.Fitter(l1_dur)
f1dur.fit()
# may take some time since by default, all distributions are tried
# but you call manually provide a smaller set of distributions
f1dur.summary()
```

# Level 2 Start Times

```python
#raw start times
l2_st = pd.read_csv('../data/L2_start_times.csv')
f_lev2_st = fitter.Fitter(l2_st)
f_lev2_st.fit()
f_lev2_st.summary()
```

```python
#raw start times
l2_dur = pd.read_csv('../data/l2_durations_xformed.csv')
f2dur = fitter.Fitter(l2_dur)
f2dur.fit()
f2dur.summary()
```

# Serializing Results

```python
with open("../data/levl1_start_times_fits.pickle", "wb") as outfile:
	pkl.dump(f_lev1_st, outfile)

with open("../data/levl2_start_times_fits.pickle", "wb") as outfile:
	pkl.dump(f_lev2_st, outfile)

with open("../data/levl1_durs_fits.pickle", "wb") as outfile:
	pkl.dump(f1dur, outfile)

with open("../data/levl2_durs_fits.pickle", "wb") as outfile:
	pkl.dump(f2dur, outfile)
```

```python
with open("../data/levl1_start_times_fits.pickle", "rb") as infile:
 	l1starts = pkl.load(infile)

with open("../data/levl2_start_times_fits.pickle", "rb") as infile:
 	l2starts = pkl.load(infile)

with open("../data/levl1_durs_fits.pickle", "rb") as infile:
 	l1durs = pkl.load(infile)

with open("../data/levl2_durs_fits.pickle", "rb") as infile:
 	l2durs = pkl.load(infile)
```
